```python
"""
Code part 1 (check schrift voor to do list)
"""

import pandas as pd
import openpyxl

###Gemeente harmonisatie

#  naamswijzigingen (zelfde gemeente, andere schrijfwijze in latere jaren)
RENAMES = {
    'Beek': 'Beek (L.)',
    'Hengelo': 'Hengelo (O.)',
    'Laren': 'Laren (NH.)',
    'Middelburg': 'Middelburg (Z.)',
    'Rijswijk': 'Rijswijk (ZH.)',
    'Stein': 'Stein (L.)',
    'Den Haag': "'s-Gravenhage",   # EV-bestand gebruikt Den Haag, KWB gebruikt 's-Gravenhage
    'Gemeente Beek': 'Beek (L.)',
    'Gemeente Hengelo': 'Hengelo (O.)',
    'Gemeente Laren': 'Laren (NH.)',
    'Gemeente Middelburg': 'Middelburg (Z.)',
    'Gemeente Rijswijk': 'Rijswijk (ZH.)',
    'Gemeente Stein': 'Stein (L.)',
    "Gemeente 's-Gravenhage": "'s-Gravenhage",
    'Gemeente Den Haag': "'s-Gravenhage",
}

# Rijen die geen echte gemeenten zijn
NON_GEMEENTE_ROWS = {
    'Nederland', 'Onbekend', 'Bron', 'Datum export',
    'klimaatmonitor.databank.nl', ''
}

# Fusies: nieuwe gemeente <- lijst van opgeheven gemeenten
MERGERS = {
    'Eemsdelta':      ['Appingedam', 'Delfzijl', 'Loppersum'],
    'Land van Cuijk': ['Boxmeer', 'Cuijk', 'Grave',
                       'Mill en Sint Hubert', 'Sint Anthonis'],
    'Maashorst':      ['Landerd', 'Uden'],
    'Dijk en Waard':  ['Heerhugowaard', 'Langedijk'],
    'Voorne aan Zee': ['Brielle', 'Hellevoetsluis', 'Westvoorne'],
}

# Absorpties: opgeheven gemeente gaat op in bestaande gemeente
ABSORPTIONS = {
    'Beemster': 'Purmerend',
    'Weesp':    'Amsterdam',
    'Haaren':   'Oisterwijk',
}

# KWB-kolommen die percentages/gemiddelden zijn: bij fusies inwoner-gewogen
# gemiddelde i.p.v. som. Alles wat hier NIET in staat wordt gesommeerd.
WEIGHTED_AVG_COLS = {
    'g_hhgro', 'bev_dich', 'a_ste', 'p_ste',
    'p_1gezw', 'p_mgezw', 'p_koopw', 'p_huurw', 'p_wcorpw',
    'g_ele', 'g_gas',
    'g_ink_po', 'g_ink_pi', 'p_ink_li', 'p_ink_hi',
    'g_hh_sti', 'p_hh_li', 'p_hh_hi',
    'p_arb_pp', 'p_arb_wn', 'p_arb_zs',
    'bev_aandeel', 'bev_phev_aandeel',
}


def harmonize_gemeenten(df: pd.DataFrame,
                        gemeente_col: str = 'gemeente') -> pd.DataFrame:

    # 1. Verwijder "Gemeente "-prefix indien aanwezig
    df[gemeente_col] = df[gemeente_col].str.replace(r'^Gemeente\s+', '',
                                                     regex=True).str.strip()
    # 2 Verwijjder metadata-rijen (geen echte gemeenten)
    df = df[~df[gemeente_col].isin(NON_GEMEENTE_ROWS)]
    df = df[df[gemeente_col].notna()]

    #3  naamswijzigingen
    df[gemeente_col] = df[gemeente_col].replace(RENAMES)

    # 3 . Fusies en absorpties per jaar aggregeren
    num_cols = df.select_dtypes(include='number').columns.tolist()
    if 'jaar' in num_cols:
        num_cols.remove('jaar')
    sum_cols = [c for c in num_cols if c not in WEIGHTED_AVG_COLS]
    avg_cols = [c for c in num_cols if c in WEIGHTED_AVG_COLS]
    weight_col = 'a_inw' if 'a_inw' in df.columns else None

    def aggregate_rows(rows: pd.DataFrame) -> dict:
        """Som voor aantallen, inwoner-gewogen gemiddelde voor %/gemiddelden."""
        out = {}
        for c in sum_cols:
            out[c] = rows[c].sum()
        if weight_col is not None and rows[weight_col].sum() > 0:
            w = rows[weight_col]
            for c in avg_cols:
                out[c] = (rows[c] * w).sum() / w.sum()
        else:
            for c in avg_cols:
                out[c] = rows[c].mean()
        return out

    result_frames = []
    for jaar, grp in df.groupby('jaar'):
        grp = grp.copy()

        for old, new in ABSORPTIONS.items():
            mask_old = grp[gemeente_col] == old
            mask_new = grp[gemeente_col] == new
            if mask_old.any() and mask_new.any():
                combined = grp[mask_old | mask_new]
                agg = pd.DataFrame([aggregate_rows(combined)])
                agg[gemeente_col] = new
                agg['jaar'] = jaar
                grp = grp[~(mask_old | mask_new)]
                grp = pd.concat([grp, agg], ignore_index=True)

        # Fusies: tel/gewogen-middel alle oude gemeenten op tot nieuwe naam
        for new_name, old_names in MERGERS.items():
            mask = grp[gemeente_col].isin(old_names)
            if mask.any():
                agg = pd.DataFrame([aggregate_rows(grp[mask])])
                agg[gemeente_col] = new_name
                agg['jaar'] = jaar
                grp = grp[~mask]
                grp = pd.concat([grp, agg], ignore_index=True)

        result_frames.append(grp)

    return pd.concat(result_frames, ignore_index=True)


# LOADER 1: EV-DATA (Y-variabele)

def load_ev(path: str = 'Autosenoverigevoertuigengemeenten.xlsx'
            ) -> pd.DataFrame:

    wb = openpyxl.load_workbook(path)
    ws = wb.active

    # Rij 2 = jaren, rij 3 = kolomnamen, rij 4+ = data
    jaren_row = [ws.cell(row=2, column=c).value for c in range(1, ws.max_column + 1)]
    cols_row  = [ws.cell(row=3, column=c).value for c in range(1, ws.max_column + 1)]

    # Bouw mapping: (jaar, kolomtype) -> kolomindex
    col_map = {
        'bev_aantal':      "Elektrische personenauto's (BEV), geregistreerd bij eigenaar [aantal]",
        'phev_aantal':     "Plug-in hybride personenauto's (PHEV), geregistreerd bij eigenaar [aantal]",
        'totaal_autos':    "Aantal personenauto's per 1 januari [aantal]",
        'bev_aandeel':     "Percentage elektrische personenauto's (BEV), geregistreerd bij eigenaar [%]",
        'bev_phev_aandeel':"Percentage elektrische personenauto's (BEV en PHEV), geregistreerd bij eigenaar [%]",
    }

    # Vind per jaar de juiste kolommen
    jaar_col_idx = {}
    current_jaar = None
    for i, (j, c) in enumerate(zip(jaren_row, cols_row)):
        if j is not None and j != '':
            try:
                current_jaar = int(j)
                jaar_col_idx[current_jaar] = {}
            except (ValueError, TypeError):
                pass
        if current_jaar and c:
            for var, label in col_map.items():
                if c.strip() == label.strip() and var not in jaar_col_idx[current_jaar]:
                    jaar_col_idx[current_jaar][var] = i  # 0-based

    records = []
    for row_idx in range(4, ws.max_row + 1):
        gemeente = ws.cell(row=row_idx, column=1).value
        if not gemeente:
            continue
        raw = [ws.cell(row=row_idx, column=c).value
               for c in range(1, ws.max_column + 1)]
        for jaar, vars_idx in jaar_col_idx.items():
            rec = {'gemeente': gemeente, 'jaar': jaar}
            for var, idx in vars_idx.items():
                val = raw[idx]
                if isinstance(val, str):
                    val = val.replace(',', '.')
                    try:
                        val = float(val)
                    except ValueError:
                        val = None
                rec[var] = val
            records.append(rec)

    df = pd.DataFrame(records)
    df = harmonize_gemeenten(df)
    df['jaar'] = df['jaar'].astype(int)
    df = df.sort_values(['gemeente', 'jaar']).reset_index(drop=True)
    return df


# LOADER 2: KWB-DATA (controlevariabelen)

# Kolommen die in alle jaargangen (2019-2024) beschikbaar zijn
KWB_COLS = [
    'gwb_code', 'gm_naam',
    'a_inw',      # aantal inwoners
    'a_00_14', 'a_15_24', 'a_25_44', 'a_45_64', 'a_65_oo',  # leeftijdsgroepen
    'a_hh',       # aantal huishoudens
    'a_1p_hh',    # eenpersoonshuishoudens
    'a_hh_m_k',   # huishoudens met kinderen
    'g_hhgro',    # gemiddelde huishoudgrootte
    'bev_dich',   # bevolkingsdichtheid (stedelijkheidsproxy)
    'a_ste',      # stedelijkheidsklasse absoluut
    'p_ste',      # stedelijkheidsklasse percentage
    'a_woning',   # aantal woningen
    'p_1gezw',    # % eengezinswoningen
    'p_mgezw',    # % meergezinswoningen
    'p_koopw',    # % koopwoningen
    'p_huurw',    # % huurwoningen
    'p_wcorpw',   # % woningcorporatie
    'g_ele',      # gem. elektriciteitsverbruik
    'g_gas',      # gem. gasverbruik
    'a_inkont',   # aantal inkomensontvangers
    'g_ink_po',   # gem. inkomen per inwoner
    'g_ink_pi',   # gem. inkomen per inkomensontvanger
    'p_ink_li',   # % lage inkomens
    'p_ink_hi',   # % hoge inkomens
    'g_hh_sti',   # gem. gestandaardiseerd huishoudinkomen
    'p_hh_li',    # % huishoudens laag inkomen
    'p_hh_hi',    # % huishoudens hoog inkomen
    'p_arb_pp',   # % arbeidspotentieel
    'p_arb_wn',   # % werknemers
    'p_arb_zs',   # % zelfstandigen
]

KWB_FILES = {
    2019: 'kwb2019.xlsx',
    2020: 'kwb2020.xlsx',
    2021: 'kwb2021.xlsx',
    2022: 'kwb2022.xlsx',
    2023: 'kwb2023.xlsx',
    2024: 'kwb2024.xlsx',
}


def _load_kwb_year(path: str, jaar: int) -> pd.DataFrame:
    wb = openpyxl.load_workbook(path, read_only=True, data_only=True)
    ws = wb.active
    header = [ws.cell(row=1, column=c).value for c in range(1, ws.max_column + 1)]
    col_idx = {h: i for i, h in enumerate(header) if h in KWB_COLS + ['recs']}

    records = []
    for row in ws.iter_rows(min_row=2, values_only=True):
        recs_val = row[col_idx.get('recs', 4)] if 'recs' in col_idx else row[4]
        if recs_val != 'Gemeente':
            continue
        # 'regio' kolom (index 2) bevat de gemeentenaam voor gemeente-rijen
        regio_idx = header.index('regio') if 'regio' in header else 2
        rec = {'jaar': jaar, 'gemeente': row[regio_idx]}
        for col in KWB_COLS:
            if col in col_idx:
                val = row[col_idx[col]]
                if isinstance(val, str):
                    val = val.strip().replace(',', '.')
                    try:
                        f = float(val)
                        val = int(f) if f.is_integer() else f
                    except ValueError:
                        val = None
                rec[col] = val
            else:
                rec[col] = None
        records.append(rec)
    return pd.DataFrame(records)


def load_kwb(data_dir: str = '.') -> pd.DataFrame:
    import os
    frames = []
    for jaar, fname in KWB_FILES.items():
        fpath = os.path.join(data_dir, fname)
        df_yr = _load_kwb_year(fpath, jaar)
        frames.append(df_yr)

    df = pd.concat(frames, ignore_index=True)
    df = harmonize_gemeenten(df)
    df['jaar'] = df['jaar'].astype(int)
    df = df.sort_values(['gemeente', 'jaar']).reset_index(drop=True)
    return df


# LOADER 3: LAADPUNTEN (infrastructuur)

def load_laadpunten(
    path: str = 'aantallenlaadpuntenpergemeente.xlsx',
    aggregatie: str = 'december'
) -> pd.DataFrame:

    wb = openpyxl.load_workbook(path)
    ws = wb.active

    # Rij 2 = maandlabels
    maand_labels = [ws.cell(row=2, column=c).value
                    for c in range(2, ws.max_column + 1)]

    from datetime import datetime
    maand_nl = {
        'januari':1,'februari':2,'maart':3,'april':4,'mei':5,'juni':6,
        'juli':7,'augustus':8,'september':9,'oktober':10,'november':11,'december':12
    }
    col_info = []
    for i, label in enumerate(maand_labels):
        if not label:
            continue
        parts = label.strip().lower().split()
        if len(parts) == 2:
            maandnr = maand_nl.get(parts[0])
            jaar = int(parts[1])
            col_info.append((jaar, maandnr, i))

    # Per gemeente: per jaar de juiste maanden pakken
    records = []
    for row_idx in range(3, ws.max_row + 1):
        gemeente = ws.cell(row=row_idx, column=1).value
        if not gemeente:
            continue
        raw = [ws.cell(row=row_idx, column=c).value
               for c in range(2, ws.max_column + 1)]

        # Groepeer per jaar
        from collections import defaultdict
        jaar_vals = defaultdict(list)
        for (jaar, maandnr, idx) in col_info:
            val = raw[idx]
            if val is not None:
                try:
                    jaar_vals[jaar].append((maandnr, float(val)))
                except (TypeError, ValueError):
                    pass

        for jaar, maand_vals in jaar_vals.items():
            if jaar < 2019 or jaar > 2024:
                continue
            if aggregatie == 'december':
                dec_vals = [v for m, v in maand_vals if m == 12]
                waarde = dec_vals[0] if dec_vals else None
            else:  # gemiddelde
                waarde = sum(v for _, v in maand_vals) / len(maand_vals) if maand_vals else None

            records.append({
                'gemeente': gemeente,
                'jaar': jaar,
                'laadpunten_totaal': waarde
            })

    df = pd.DataFrame(records)
    df = harmonize_gemeenten(df)
    df['jaar'] = df['jaar'].astype(int)
    df = df.sort_values(['gemeente', 'jaar']).reset_index(drop=True)
    return df


# SEPP-SUBSIDIE (tijdsvariabele)

# Subsidiebedragen particulieren per jaar
SEPP_BEDRAGEN = {
    2019: 0,
    2020: 4000,
    2021: 4000,
    2022: 3350,
    2023: 2950,
    2024: 2550,
}

def add_sepp(df: pd.DataFrame, jaar_col: str = 'jaar') -> pd.DataFrame:
    df = df.copy()
    df['sepp_actief'] = df[jaar_col].map(
        lambda j: 1 if SEPP_BEDRAGEN.get(j, 0) > 0 else 0)
    df['sepp_bedrag'] = df[jaar_col].map(
        lambda j: SEPP_BEDRAGEN.get(j, 0))
    return df


# QUICK TEST

if __name__ == '__main__':
    import os
    DATA = '/content'

    print('=== load_ev ===')
    ev = load_ev(os.path.join(DATA, 'Autosenoverigevoertuigengemeenten.xlsx'))
    print(ev.shape, '\n', ev.head(3).to_string())

    print('\n=== load_kwb ===')
    kwb = load_kwb(DATA)
    print(kwb.shape, '\n', kwb[['gemeente','jaar','a_inw','p_koopw','g_ink_pi']].head(3).to_string())

    print('\n=== load_laadpunten ===')
    laad = load_laadpunten(os.path.join(DATA, 'aantallenlaadpuntenpergemeente.xlsx'))
    print(laad.shape, '\n', laad.head(3).to_string())

    print('\n=== merge check ===')
    merged = (ev[ev.jaar.between(2019,2024)]
              .merge(kwb, on=['gemeente','jaar'], how='inner')
              .merge(laad, on=['gemeente','jaar'], how='inner'))
    merged = add_sepp(merged)
    print('Merged shape:', merged.shape)
    print('Verwacht: ~2052 rijen (342 gem x 6 jaar)')
    print(merged[['gemeente','jaar','bev_phev_aandeel','g_ink_pi','laadpunten_totaal','sepp_bedrag']].head(6).to_string())
```

```python
"""
regression
"""

import numpy as np
import pandas as pd
from scipy import stats

from loaders import load_ev, load_kwb, load_laadpunten, add_sepp


# 1. DATASET OPBOUWEN

def build_dataset(data_dir: str = '.') -> pd.DataFrame:

    ev = load_ev(f'{data_dir}/Autosenoverigevoertuigengemeenten.xlsx')
    kwb = load_kwb(data_dir)
    laad = load_laadpunten(f'{data_dir}/aantallenlaadpuntenpergemeente.xlsx')

    df = (ev[ev.jaar.between(2019, 2024)]
          .merge(kwb, on=['gemeente', 'jaar'], how='inner')
          .merge(laad, on=['gemeente', 'jaar'], how='inner'))
    df = add_sepp(df)

    # Afgeleide variabelen
    df['laadpunten_per_1000inw'] = df['laadpunten_totaal'] / df['a_inw'] * 1000
    df['log_laadpunten'] = np.log1p(df['laadpunten_totaal'])
    df['log_inkomen'] = np.log(df['g_ink_pi'])
    df['log_bevdich'] = np.log(df['bev_dich'])

    # Gelagde laadinfrastructuur (t-1) tegen reverse causality
    # naar jaar t+1, zodat het de "vorig-jaar" waarde wordt voor jaar t+1.
    lag = laad[['gemeente', 'jaar', 'laadpunten_totaal']].copy()
    inw = df[['gemeente', 'jaar', 'a_inw']].drop_duplicates()
    lag = lag.merge(inw, on=['gemeente', 'jaar'], how='left')
    lag['laadpunten_per_1000inw_lag'] = lag['laadpunten_totaal'] / lag['a_inw'] * 1000
    lag['log_laadpunten_lag'] = np.log1p(lag['laadpunten_totaal'])
    lag['jaar'] = lag['jaar'] + 1  # waarde van jaar t wordt t-1 voor jaar t+1
    lag = lag[['gemeente', 'jaar', 'laadpunten_per_1000inw_lag', 'log_laadpunten_lag']]

    df = df.merge(lag, on=['gemeente', 'jaar'], how='left')

    # Provincie toevoegen (voor provincie x jaar fixed effects)
    from provincie_mapping import PROVINCIE
    df['provincie'] = df['gemeente'].map(PROVINCIE)

    return df


# 2. DESCRIPTIEVE STATISTIEKEN

def describe(df: pd.DataFrame, cols=None) -> pd.DataFrame:
    """Tabel met N, gemiddelde, std, min, max per variabele."""
    if cols is None:
        cols = ['bev_aandeel', 'bev_phev_aandeel', 'laadpunten_totaal',
                'laadpunten_per_1000inw', 'g_ink_pi', 'p_koopw',
                'p_1gezw', 'bev_dich', 'a_inw', 'sepp_bedrag']
    desc = df[cols].describe().T
    desc = desc[['count', 'mean', 'std', 'min', 'max']]
    desc.columns = ['N', 'Gemiddelde', 'Std.dev.', 'Min', 'Max']
    return desc.round(3)


# 3. OLS-SCHATTER

def ols(y: np.ndarray, X: np.ndarray, varnames, robust: bool = True,
        cluster: np.ndarray = None):
    """
      Schat OLS coefficienten met standaardfouten.
    """
    n, k = X.shape
    X1 = np.column_stack([np.ones(n), X])
    names = ['const'] + list(varnames)
    k1 = k + 1

    XtX_inv = np.linalg.inv(X1.T @ X1)
    beta = XtX_inv @ X1.T @ y
    resid = y - X1 @ beta
    dof = n - k1

    if cluster is not None:
        # Cluster-robust covariance
        meat = np.zeros((k1, k1))
        clusters = np.unique(cluster)
        G = len(clusters)
        for c in clusters:
            mask = cluster == c
            Xg = X1[mask]
            ug = resid[mask]
            score = Xg.T @ ug
            meat += np.outer(score, score)
        adj = (G / (G - 1)) * ((n - 1) / dof)
        cov = adj * XtX_inv @ meat @ XtX_inv
    elif robust:
        # HC1 robust covariance
        meat = (X1 * resid[:, None]).T @ (X1 * resid[:, None])
        cov = (n / dof) * XtX_inv @ meat @ XtX_inv
    else:
        sigma2 = (resid @ resid) / dof
        cov = sigma2 * XtX_inv

    se = np.sqrt(np.diag(cov))
    tstat = beta / se
    pval = 2 * (1 - stats.t.cdf(np.abs(tstat), dof))
    crit = stats.t.ppf(0.975, dof)
    ci_low = beta - crit * se
    ci_high = beta + crit * se

    # R-squared
    ss_res = resid @ resid
    ss_tot = ((y - y.mean()) ** 2).sum()
    r2 = 1 - ss_res / ss_tot
    r2_adj = 1 - (1 - r2) * (n - 1) / dof

    result = pd.DataFrame({
        'coef': beta,
        'std err': se,
        't': tstat,
        'p-value': pval,
        '95% CI low': ci_low,
        '95% CI high': ci_high,
    }, index=names)

    meta = {'n_obs': n, 'r2': r2, 'r2_adj': r2_adj, 'dof': dof}
    return result.round(4), meta


# 4. FIXED-EFFECTS (within-transformatuie: gemeente + JAar dumies)

def fixed_effects(df: pd.DataFrame, y_col: str, x_cols: list,
                   entity_col: str = 'gemeente', time_col: str = 'jaar'):
    """
    Two-way fixed-effects model
    """
    data = df[[entity_col, time_col, y_col] + x_cols].dropna().copy()

    # Within-transformatie
    def demean(group_col, cols):
        means = data.groupby(group_col)[cols].transform('mean')
        return data[cols] - means

    overall_mean = data[[y_col] + x_cols].mean()

    # Stap 1: trek entity-gemiddelden af
    entity_demeaned = data[[y_col] + x_cols] - data.groupby(entity_col)[[y_col] + x_cols].transform('mean')

    temp = entity_demeaned.copy()
    temp[entity_col] = data[entity_col].values
    temp[time_col] = data[time_col].values
    time_means = temp.groupby(time_col)[[y_col] + x_cols].transform('mean')
    two_way = entity_demeaned - time_means

    two_way = two_way + overall_mean

    y = two_way[y_col].values
    X = two_way[x_cols].values
    cluster = data[entity_col].values

    result, meta = ols(y, X, x_cols, robust=True, cluster=cluster)

    n_entities = data[entity_col].nunique()
    n_times = data[time_col].nunique()
    meta['dof_fe'] = meta['n_obs'] - len(x_cols) - 1 - (n_entities - 1) - (n_times - 1)
    meta['n_entities'] = n_entities
    meta['n_times'] = n_times

    result = result.drop(index='const')
    return result, meta


def fixed_effects_with_trend(df: pd.DataFrame, y_col: str, x_cols: list,
                              entity_col: str = 'gemeente',
                              time_col: str = 'jaar'):
    """
    Two-way fixed-effects model UITGEBREID
    """
    data = df[[entity_col, time_col, y_col] + x_cols].dropna().copy()
    data = data.reset_index(drop=True)

    # Centreer jaar
    jaar_mean = data[time_col].mean()
    data['_t'] = data[time_col] - jaar_mean

    entities = data[entity_col].unique()

    def partial_out(var_col):
        ####Residuen van var_col na regressie op:
           ###gemeente-dummies + gemeente-specifieke trend (gemeente x _t) + jaar-dummies
        y = data[var_col].values.astype(float)
        resid = y.copy()

        # Stap 1: voor elke gemeente, trek de eigen lineaire trend in _t eraf
        for e in entities:
            mask = (data[entity_col] == e).values
            t_e = data.loc[mask, '_t'].values
            y_e = resid[mask]
            if len(t_e) < 2 or np.var(t_e) == 0:
                resid[mask] = y_e - y_e.mean()
                continue
            # OLS van y_e op [1, t_e]
            T = np.column_stack([np.ones(len(t_e)), t_e])
            coef = np.linalg.lstsq(T, y_e, rcond=None)[0]
            resid[mask] = y_e - T @ coef

        # Stap 2: trek jaar-gemiddelden van de residuen eraf (jaar-FE)
        resid_series = pd.Series(resid, index=data.index)
        jaar_means = resid_series.groupby(data[time_col]).transform('mean')
        resid_final = resid_series - jaar_means

        return resid_final.values

    y_tilde = partial_out(y_col)
    X_tilde = np.column_stack([partial_out(c) for c in x_cols])

    cluster = data[entity_col].values
    result, meta = ols(y_tilde, X_tilde, x_cols, robust=True, cluster=cluster)

    n_entities = data[entity_col].nunique()
    n_times = data[time_col].nunique()
    # Effectief aantal parameters: k (x_cols) + 2*n_entities (intercept+trend per gemeente) + (n_times-1) jaardummies
    n_params = len(x_cols) + 2 * n_entities + (n_times - 1)
    meta['dof_fe'] = meta['n_obs'] - n_params
    meta['n_entities'] = n_entities
    meta['n_times'] = n_times
    meta['n_params'] = n_params

    result = result.drop(index='const')
    return result, meta


def fixed_effects_province_year(df: pd.DataFrame, y_col: str, x_cols: list,
                                 entity_col: str = 'gemeente',
                                 time_col: str = 'jaar',
                                 provincie_col: str = 'provincie'):

    ######Fixed-effects model met gemeente-FE + provincie x jaar-FE.$####

    data = df[[entity_col, time_col, provincie_col, y_col] + x_cols].dropna().copy()
    data = data.reset_index(drop=True)

    cols = [y_col] + x_cols
    overall_mean = data[cols].mean()

    # Stap 1: gemeente-demeaning
    entity_demeaned = data[cols] - data.groupby(entity_col)[cols].transform('mean')

    # Stap 2: jaar-demeaning (op entity-demeaned data)
    temp = entity_demeaned.copy()
    temp[time_col] = data[time_col].values
    jaar_means = temp.groupby(time_col)[cols].transform('mean')
    two_way = entity_demeaned - jaar_means

    # Stap 3: provincie x jaar-demeaning (op de two-way-residuen)
    temp2 = two_way.copy()
    temp2['_pj'] = data[provincie_col].astype(str) + '_' + data[time_col].astype(str)
    pj_means = temp2.groupby('_pj')[cols].transform('mean')
    three_way = two_way - pj_means

    # overall mean terugplakken zodat schaal/interpretatie klopt
    final = three_way + overall_mean

    y = final[y_col].values
    X = final[x_cols].values
    cluster = data[entity_col].values

    result, meta = ols(y, X, x_cols, robust=True, cluster=cluster)

    n_entities = data[entity_col].nunique()
    n_times = data[time_col].nunique()
    n_prov = data[provincie_col].nunique()
    # parameters: k + (n_entities-1) gemeente-FE + (n_times-1) jaar-FE
    #             + (n_prov-1)*(n_times-1) extra voor provincie x jaar interactie
    n_params = (len(x_cols) + (n_entities - 1) + (n_times - 1)
                + (n_prov - 1) * (n_times - 1))
    meta['dof_fe'] = meta['n_obs'] - n_params
    meta['n_entities'] = n_entities
    meta['n_times'] = n_times
    meta['n_provincies'] = n_prov
    meta['n_params'] = n_params

    result = result.drop(index='const')
    return result, meta


# 5. MAIN: BEIDE MODELLEN VOOR BEIDE Y-VARIABELEN

# Set van verklarende variabelen voor het hoofdmodel (gelijktijdig)
# Hoofdmodel (met SEPP)
X_VARS = [
    'laadpunten_per_1000inw',
    'g_ink_pi',
    'p_koopw',
    'p_1gezw',
    'bev_dich',
    'sepp_bedrag',
]

# Robustness check (zonder SEPP)
X_VARS_NO_SEPP = [
    'laadpunten_per_1000inw',
    'g_ink_pi',
    'p_koopw',
    'p_1gezw',
    'bev_dich',
]

# Lag-model met SEPP
X_VARS_LAG = [
    'laadpunten_per_1000inw_lag',
    'g_ink_pi',
    'p_koopw',
    'p_1gezw',
    'bev_dich',
    'sepp_bedrag',
]

# Lag-model zonder SEPP
X_VARS_LAG_NO_SEPP = [
    'laadpunten_per_1000inw_lag',
    'g_ink_pi',
    'p_koopw',
    'p_1gezw',
    'bev_dich',
]

def run_all(data_dir: str = '.'):
    df = build_dataset(data_dir)

    print('=' * 70)
    print('DESCRIPTIEVE STATISTIEKEN (N =', len(df), ')')
    print('=' * 70)
    print(describe(df))

    results = {}
    for y_col, label in [('bev_aandeel', 'BEV (hoofdmaat)'),
                          ('bev_phev_aandeel', 'BEV+PHEV (robuustheid)')]:
        print('\n' + '=' * 70)
        print(f'Y = {y_col}  [{label}]')
        print('=' * 70)

        data = df.dropna(subset=[y_col] + X_VARS)

        print(f'\n--- Pooled OLS, gelijktijdig (n={len(data)}) ---')
        pooled, meta_p = ols(data[y_col].values, data[X_VARS].values,
                              X_VARS, robust=True)
        print(pooled)
        print(f"R²={meta_p['r2']:.4f}  R²_adj={meta_p['r2_adj']:.4f}  n={meta_p['n_obs']}")

        print(f'\n--- Fixed Effects, gelijktijdig (gemeente + jaar) ---')
        fe, meta_fe = fixed_effects(data, y_col, X_VARS)
        print(fe)
        print(f"R²(within)={meta_fe['r2']:.4f}  n={meta_fe['n_obs']}  "
              f"gemeenten={meta_fe['n_entities']}  jaren={meta_fe['n_times']}")

        # Lag-model: laadpunten van t-1 (verzacht reverse causality)
        data_lag = df.dropna(subset=[y_col] + X_VARS_LAG)
        print(f'\n--- Fixed Effects, laadpunten t-1 (n={len(data_lag)}) ---')
        print('(2019 valt weg: laadpaaldata start medio 2019, geen t-1 beschikbaar)')
        fe_lag, meta_fe_lag = fixed_effects(data_lag, y_col, X_VARS_LAG)
        print(fe_lag)
        print(f"R²(within)={meta_fe_lag['r2']:.4f}  n={meta_fe_lag['n_obs']}  "
              f"gemeenten={meta_fe_lag['n_entities']}  jaren={meta_fe_lag['n_times']}")

  #######################

        print(f'\n--- Fixed Effects + provincie x jaar, laadpunten t-1 ---')
        fe_prov, meta_fe_prov = fixed_effects_province_year(
            data_lag,
            y_col,
            X_VARS_LAG
        )

        print(fe_prov)

        print(
            f"R²={meta_fe_prov['r2']:.4f}  "
            f"n={meta_fe_prov['n_obs']}  "
            f"provincies={meta_fe_prov['n_provincies']}  "
            f"parameters={meta_fe_prov['n_params']}  "
            f"dof={meta_fe_prov['dof_fe']}"
        )

        # ROBUSTNESS CHECK: subside uitsliuten

        print('\n--- ROBUSTNESS: FE WITHOUT SEPP ---')

        data_nosepp = df.dropna(
            subset=[y_col] + X_VARS_NO_SEPP
        )

        fe_nosepp, meta_nosepp = fixed_effects(
            data_nosepp,
            y_col,
            X_VARS_NO_SEPP
        )

        print(fe_nosepp)

        print(
            f"R²(within)={meta_nosepp['r2']:.4f}  "
            f"n={meta_nosepp['n_obs']}"
        )

        print('\n--- ROBUSTNESS: FE LAGGED WITHOUT SEPP ---')

        data_lag_nosepp = df.dropna(
            subset=[y_col] + X_VARS_LAG_NO_SEPP
        )

        fe_lag_nosepp, meta_lag_nosepp = fixed_effects(
            data_lag_nosepp,
            y_col,
            X_VARS_LAG_NO_SEPP
        )

        print(fe_lag_nosepp)

        print(
            f"R²(within)={meta_lag_nosepp['r2']:.4f}  "
            f"n={meta_lag_nosepp['n_obs']}"
        )

        results[y_col] = {
            'pooled': (pooled, meta_p),
            'fe': (fe, meta_fe),
            'fe_lag': (fe_lag, meta_fe_lag),
            'fe_prov': (fe_prov, meta_fe_prov),
            'fe_nosepp': (fe_nosepp, meta_nosepp),
            'fe_lag_nosepp': (fe_lag_nosepp, meta_lag_nosepp)
        }

    return df, results

if __name__ == '__main__':
    df, results = run_all('.')

    # Sla het analysebestand op
    df.to_csv('analysebestand_2019_2024.csv', index=False)
    print('\nAnalysebestand opgeslagen als analysebestand_2019_2024.csv')


```

```python
# TABLE 3: MAIN REGRESSION RESULTS

import pandas as pd

def make_main_results_table(results, y_col='bev_aandeel'):

    # Models
    pooled, meta_p = results[y_col]['pooled']
    fe, meta_fe = results[y_col]['fe']
    fe_lag, meta_fe_lag = results[y_col]['fe_lag']
    fe_prov, meta_fe_prov = results[y_col]['fe_prov']

    models = {
        '(1) Pooled OLS': pooled,
        '(2) Fixed Effects': fe,
        '(3) Fixed Effects (Lagged Infrastructure)': fe_lag,
        '(4) Fixed Effects + Province-Year FE': fe_prov
    }

    # Variables to display
    variables = [
        'laadpunten_per_1000inw',
        'laadpunten_per_1000inw_lag',
        'g_ink_pi',
        'p_koopw',
        'p_1gezw',
        'bev_dich',
        'sepp_bedrag'
    ]

    # English labels
    labels = {
        'laadpunten_per_1000inw': 'Charging Points per 1,000 Inhabitants',
        'laadpunten_per_1000inw_lag': 'Charging Points per 1,000 Inhabitants (t−1)',
        'g_ink_pi': 'Disposable Income per Capita',
        'p_koopw': 'Owner-Occupied Housing (%)',
        'p_1gezw': 'Single-Family Housing (%)',
        'bev_dich': 'Population Density',
        'sepp_bedrag': 'SEPP Subsidy (€)'
    }

    rows = []

    for var in variables:

        row = {'Variable': labels[var]}

        for model_name, model in models.items():

            if var not in model.index:
                row[model_name] = ''
                continue

            coef = model.loc[var, 'coef']
            pval = model.loc[var, 'p-value']

            if pval < 0.01:
                stars = '***'
            elif pval < 0.05:
                stars = '**'
            elif pval < 0.10:
                stars = '*'
            else:
                stars = ''

            row[model_name] = f'{coef:.4f}{stars}'

        rows.append(row)

    # Create dataframe
    table = pd.DataFrame(rows)
    table = table.set_index('Variable')

    # Model statistics
    table.loc['Observations'] = [
        meta_p['n_obs'],
        meta_fe['n_obs'],
        meta_fe_lag['n_obs'],
        meta_fe_prov['n_obs']
    ]

    table.loc['R²'] = [
        round(meta_p['r2'], 3),
        round(meta_fe['r2'], 3),
        round(meta_fe_lag['r2'], 3),
        round(meta_fe_prov['r2'], 3)
    ]

    table.loc['Municipality Fixed Effects'] = [
        'No', 'Yes', 'Yes', 'Yes'
    ]

    table.loc['Year Fixed Effects'] = [
        'No', 'Yes', 'Yes', 'Yes'
    ]

    table.loc['Province-Year Fixed Effects'] = [
        'No', 'No', 'No', 'Yes'
    ]

    return table


# CREATE TABLE

main_table = make_main_results_table(results)

# DISPLAY TABLE

print("\n")
print("=" * 100)
print("TABLE 3. MAIN REGRESSION RESULTS")
print("=" * 100)
print(main_table.to_string())

print("\nNotes:")
print("*** p < 0.01, ** p < 0.05, * p < 0.10")
print("Dependent variable: Battery Electric Vehicle (BEV) Share")

# EXPORT FILES

xlsx_file = "Table_3_Main_Regression_Results.xlsx"
csv_file = "Table_3_Main_Regression_Results.csv"

main_table.to_excel(
    xlsx_file,
    sheet_name="Main Results"
)

main_table.to_csv(
    csv_file
)

print("\nFiles exported successfully:")
print(f"- {xlsx_file}")
print(f"- {csv_file}")

# AUTOMATIC DOWNLOAD

try:
    from google.colab import files

    files.download(xlsx_file)
    files.download(csv_file)

except:
    print("\nNot running in Google Colab.")
    print("Download files manually from the file browser.")
```

```python
import pandas as pd
from IPython.display import display

def make_robustness_table(results):

    pooled, meta_p = results['bev_aandeel']['pooled']
    fe, meta_fe = results['bev_aandeel']['fe']
    fe_lag, meta_fe_lag = results['bev_aandeel']['fe_lag']
    fe_prov, meta_fe_prov = results['bev_aandeel']['fe_prov']

    def coef_string(model, var):

        if var not in model.index:
            return ""

        coef = model.loc[var, 'coef']
        p = model.loc[var, 'p-value']

        if p < 0.01:
            stars = "***"
        elif p < 0.05:
            stars = "**"
        elif p < 0.10:
            stars = "*"
        else:
            stars = ""

        return f"{coef:.4f}{stars}"

    table = pd.DataFrame({
        'FE': [
            coef_string(fe, 'laadpunten_per_1000inw'),
            '',
            coef_string(fe, 'sepp_bedrag'),
            meta_fe['n_obs'],
            round(meta_fe['r2'], 3)
        ],
        'FE Lagged': [
            '',
            coef_string(fe_lag, 'laadpunten_per_1000inw_lag'),
            coef_string(fe_lag, 'sepp_bedrag'),
            meta_fe_lag['n_obs'],
            round(meta_fe_lag['r2'], 3)
        ],
        'FE + Province×Year FE': [
            '',
            coef_string(fe_prov, 'laadpunten_per_1000inw_lag'),
            coef_string(fe_prov, 'sepp_bedrag'),
            meta_fe_prov['n_obs'],
            round(meta_fe_prov['r2'], 3)
        ]
    },
    index=[
        'Charging Stations',
        'Charging Stations (t-1)',
        'SEPP Subsidy',
        'Observations',
        'R²'
    ])

    return table


# Create table
rob_table = make_robustness_table(results)

print("\n" + "="*80)
print("TABLE 4. ROBUSTNESS CHECKS")
print("="*80)

display(rob_table)

# Save Excel file
filename = "Table4_Robustness_Checks.xlsx"
rob_table.to_excel(filename)

print(f"\nSaved as: {filename}")

# Automatic download in Google Colab
try:
    from google.colab import files
    files.download(filename)
except:
    print("Download manually from the file browser.")
```
